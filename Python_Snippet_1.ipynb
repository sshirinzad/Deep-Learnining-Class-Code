#Problem 3

```python
import numpy as np

def sigmoid(z):
    result = 1 / (1 + np.exp(-z))
    return result

# Testing the function with different inputs
print(sigmoid(0))   # Output should be 0.5
print(sigmoid(2))   # Output should be close to 0.88
print(sigmoid(-2))  # Output should be close to 0.12
```

#Problem 4


```python
import numpy as np
import matplotlib.pyplot as plt



def tanh(z):
    return np.sinh(z) / np.cosh(z)

# Testing the function with different inputs
print(tanh(0))  # Expected output: 0
print(tanh(1))   # Expected output: 0.7616...
print(tanh(-1))  # Expected output: -0.7616...
print(tanh(2))   # Expected output: 0.9640...
print(tanh(-2))  # Expected output: -0.9640...

# Bonus: Plotting the hyperbolic tangent function
z_values = np.linspace(-5, 5, 100)
tanh_values = tanh(z_values)

plt.plot(z_values, tanh_values)
plt.title('Hyperbolic Tangent Function')
plt.xlabel('z')
plt.ylabel('tanh(z)')
plt.grid(True)
plt.show()
```

#Problem 5


```python
import numpy as np

def relu(z):
    return np.maximum(0, z)

# Testing the function with a scalar, a list, and a numpy array
print(relu(-3))                  # Scalar input, Expected output: 0
print(relu([0, -1, 2, -3, 4]))   # List input, Expected output: [0, 0, 2, 0, 4]
print(relu(np.array([-3, -2, 0, 1, 2])))  # Numpy array input, Expected output: [0, 0, 0, 1, 2]
```
